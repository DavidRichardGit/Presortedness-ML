## Sample Size Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Comparisons needed to compute the presortedness values

```python
sample_sizes = range(1, 31)

avg_presortedness_comparisons = [
    2.3, 4.7, 7.4, 10.1, 13.0, 16.1, 19.2, 22.4, 25.8,
    29.2, 32.6, 36.2, 39.7, 43.3, 47.0, 50.7, 54.6, 58.4, 62.2,
    64.6, 68.6, 72.7, 77.0, 81.0, 85.3, 89.5, 93.9, 98.2, 102.5, 
    106.9
]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, avg_presortedness_comparisons, marker='o', color='blue')

# Labels and title
plt.xlabel('Sample Size')
plt.ylabel('Average Comparisons')
plt.title('Average Comparisons Needed to Compute Presortedness')
plt.grid(True)

# Show plot
plt.show()
```

## Optimal sample size at different sequence sizes

```python
array_sizes = [200,400,600,800,1000]

optimal_sample_size = [13,22,38,45,44]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(array_sizes, optimal_sample_size, marker='o', color='blue')

# Labels and title
plt.xlabel('Sequence Size')
plt.ylabel('Optimal Sample Size')
plt.title('Optimal Sample Size by Sequence Size')
plt.grid(True)

# Show plot
plt.show()
```
